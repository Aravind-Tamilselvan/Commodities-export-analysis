Commodities Exported from APRIL 2020 - APRIL 2021

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('../excel/commodities 2020-2021.xlsx')
df
```

```python
# Change datatype of S.NO and HSCode
df = df.astype({'S.No.':'Int64','HSCode':'Int64'})
df
```

Removing duplicate columns

```python
#df.T.duplicated  df.T => swaps rows to columns and columns to rows
#df.T.duplicated => check for duplicate values in rows return as True
#~df.T.duplicated() negates false to true by which non duplicate columns becomes true
#df.loc[:,~df.T.duplicated()] this selects all the non duplicate columns

new_df = df.loc[:,~df.T.duplicated()]
new_df
```

```python
#replacing \n with space
new_df.loc[:, 'Commodity'] = new_df['Commodity'].str.replace('\n',' ',regex=False)
new_df.columns = new_df.columns.str.replace('\n',' ',regex=False)
new_df
```

```python
new_df.to_excel('commodities 2020-21.xlsx',index=False)
```
