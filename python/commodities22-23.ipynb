Commodities exported from APRIL 2022- APRIL 2023

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = pd.read_excel('../excel/commodities 2022-2023.xlsx')
df
```

```python
# Change datatype of S.NO and HSCode
df = df.astype({'S.No.':'Int64','HSCode':'Int64'})
df
```

```python
new_df = df.loc[:,~df.T.duplicated()]
new_df
```

```python
new_df.loc[:, 'Commodity'] = new_df['Commodity'].str.replace('\n',' ',regex=False)
new_df.columns = new_df.columns.str.replace('\n',' ',regex=False)
new_df
```

```python
new_df.to_excel('commodities 2022-23.xlsx',index=False)
```
